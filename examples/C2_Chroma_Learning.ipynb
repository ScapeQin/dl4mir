```python
# Standard Imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import mpld3
import seaborn
import json

import optimus
import biggie

seaborn.set()
mpld3.enable_notebook()
np.set_printoptions(precision=4)

import dl4mir.chords.data as D
import dl4mir.common.streams as S
import dl4mir.common.util as util
```

```python
# Use the CQT features from Chapter 5: Large Vocabulary Automatic Chord Estimation
stash = biggie.Stash(
    "/Volumes/megatron/dl4mir/chord_estimation/biggie/chords_l2n/0/train.hdf5", 
    cache=True)
```

```python
stream = D.create_target_stream(stash, 1)
batch = S.minibatch(stream, batch_size=50)
```

```python
# Inputs
input_data = optimus.Input(name="data", shape=(None, 1, 1, 252))
target = optimus.Input(name="target", shape=(None, 12))
learning_rate = optimus.Input(name="learning_rate", shape=None)

# Single linear dot-product layer
affine = optimus.Affine(
    "affine", input_shape=(None, 1, 1, 252), 
    output_shape=(None, 12), act_type='linear')
mse = optimus.MeanSquaredErrorLoss("mse")
loss = optimus.Output("loss")
edges = optimus.ConnectionManager([
    (input_data, affine.input),
    (affine.output, mse.prediction),
    (target, mse.target),
    (mse.output, loss)])

# Only connect the learning rate to the weights (not the bias).
updates = optimus.ConnectionManager([
    (learning_rate, affine.weights)])

optimus.random_init(affine.weights)

trainer = optimus.Graph(
    "trainer",
    inputs=[input_data, target, learning_rate],
    nodes=[affine, mse],
    connections=edges.connections,
    outputs=[loss],
    updates=updates.connections,
    loss=loss)

chroma = optimus.Output("chroma")
edges = optimus.ConnectionManager([
    (input_data, affine.input),
    (affine.output, chroma)])

predictor = optimus.Graph(
    "predictor",
    inputs=[input_data],
    nodes=[affine],
    connections=edges.connections,
    outputs=[chroma])
```

```python
driver = optimus.Driver(graph=trainer)
```

```python
driver.fit(batch, {"learning_rate": 0.001}, print_freq=5000, max_iter=50000)
```
