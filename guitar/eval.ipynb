## Pulling in Results

Here, we'll use the same partitions as in the previous analysis of chord estimation results.

```python
# Configuration parameter setup.
res_fmt = "{basedir}/results/{model}/0.125/{fold}/{split}/chords/{partition}.json"

models = ['XL', 'XL_weighted_asis', 'XL_weighted']
folds = range(5)
splits = ['train', 'valid', 'test']
partitions = ['all', 'strict', 'other']
metrics = ['triads', 'root', 'v157_strict', 'mirex', 
           'tetrads', 'sevenths', 'thirds', 'majmin']
# Sorted by occurence in the data
quality_labels = ['maj', 'min', '7', 'min7', 'N', 'maj7', 'maj6',
                  'sus4', 'sus2', 'aug', 'dim', 'min6','hdim7', 'dim7']
```

```python
# Collect results over a model.
def collect_results(model_idx, partition_idx=1):
    macro_aves = np.zeros([3, 5])
    micro_recall = np.zeros([3, 5, 8])
    qualitywise = np.zeros([3, 14, 2, 5])
    for si, s in enumerate(splits):
        for fi, f in enumerate(folds):
            res = json.load(open(res_fmt.format(
                basedir=guitar_dir, model=models[model_idx],
                fold=f, split=s, partition=partitions[partition_idx])))
            # v157 is the third position in the list
            macro_aves[si, fi] = np.mean(res['macro_average'][2][1])
            these_quals = res['macro_average'][2][0]
            qualitywise[si, :, 0, fi] = np.array(
                [res['macro_average'][2][1][these_quals.index(q)] for q in quals])
            qualitywise[si, :, 1, fi] = np.array(
                [res['macro_average'][2][2][these_quals.index(q)] for q in quals])
            scores, supports = [np.array(_) for _ in res['score_annotations']]
            scalar = supports.sum(axis=0)
            scalar[scalar == 0] = 1.0
            micro_recall[si, fi, :] = (supports * scores).sum(axis=0) / scalar
    return macro_aves, micro_recall, qualitywise
```

```python
# Pretty-printer for overall stats
def print_microrecall(micro_recall, tablefmt='plain'):
    res = micro_recall.mean(axis=1).tolist()
    print tabulate.tabulate([[s] + row for s, row in zip(splits, res)], 
                            headers=[""] + metrics, floatfmt='.4f', tablefmt=tablefmt)
```

```python
# Pretty-printer for qualitywise stats
def print_qualitywise(qualitywise, tablefmt='plain'):
    table = np.concatenate([_ for _ in qualitywise.mean(axis=-1)], axis=-1)
    table[:, 1::2] /= 60.0
    table = table.tolist() + [table.mean(axis=0).tolist()]
    print tabulate.tabulate([[label] + row 
                             for label, row in zip(quality_labels + ['total'], table)], 
                            headers=['quality'] + ['recall', 'support'] * 3, 
                            floatfmt=".4f", tablefmt=tablefmt)
```

## Error Analysis

Beyond regularizing / constraining the learning problem, an additional argument for the fretboard representation is to make prediction error more informative. Hard classification systems ---those that output a single concept--- can be viewed as introducing quantization error at the output. Here, we can compare the regression output against the quantized fretboard representation to see what is gained or lost in the process.

To do this, we'll histogram the distances between the fretboard representation and its target template, and compare this to the  distances between templates of the predicted and reference templates.

```python
templates = np.load("/Volumes/megatron/dl4mir/guitar/v157_targets.npy")
```

```python
def ideal_distances(entity):
    class_idx = vocab.label_to_index(entity.chord_labels)
    valid_idx = np.array([_ is not None for _ in class_idx])
    ref_frets = templates[class_idx[valid_idx].astype(int), ...]
    est_frets = templates[entity.posterior[valid_idx].argmax(axis=1)]
    return np.sqrt(np.power(ref_frets - est_frets, 2.0).reshape(-1, 54).sum(axis=1))

def fret_distances(entity):
    class_idx = vocab.label_to_index(entity.chord_labels)
    valid_idx = np.array([_ is not None for _ in class_idx])
    ref_frets = templates[class_idx[valid_idx].astype(int), ...]
    return np.sqrt(np.power(ref_frets - entity.fretboard[valid_idx], 2.0).reshape(-1, 54).sum(axis=1))
```

```python
quantized_dists = []
soft_dists_true = []
soft_dists_false = []
post_fmt = "/Volumes/megatron/dl4mir/guitar/outputs/XL_weighted/0.125/{0}/posteriors/test.hdf5"
fret_fmt = "/Volumes/megatron/dl4mir/guitar/outputs/XL_weighted/0.125/{0}/fretboard/test.hdf5"
for n in folds:
    post_stash = biggie.Stash(post_fmt.format(n))
    fret_stash = biggie.Stash(fret_fmt.format(n))
    for k in post_stash.keys():
        qdist = ideal_distances(post_stash.get(k))
        tf_idx = np.equal(qdist, 0)
        quantized_dists.extend(qdist.tolist())
        frets = fret_stash.get(k)
        sdist = fret_distances(frets)
        soft_dists_true.extend(sdist[tf_idx].tolist())
        soft_dists_false.extend(sdist[np.invert(tf_idx)].tolist())
    print "Finished {}".format(n)
```

```python
dmax = np.sqrt(12)
quant_bins, quant_edges = np.histogram(quantized_dists, bins=50, range=(0, dmax))
soft_true_bins, soft_edges = np.histogram(soft_dists_true, bins=100, range=(0, dmax))
soft_false_bins = np.histogram(soft_dists_false, bins=100, range=(0, dmax))[0]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))
qwidth = np.abs(np.diff(quant_edges)).mean()
axes[0].bar(quant_edges[:1], quant_bins[:1], width=qwidth, fc=[0.3, 0.65, 0.3])
axes[0].bar(quant_edges[1:-1], quant_bins[1:], width=qwidth, fc=[0.3, 0.3, .75])
axes[1].plot(soft_edges[:-1], soft_true_bins, c=[0.3, 0.65, 0.3])
axes[1].plot(soft_edges[:-1], soft_false_bins, c=[0.3, 0.3, .75])
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.gca()
qwidth = np.abs(np.diff(quant_edges)).mean()
total_count = float(np.sum(quant_bins))
ax.bar(quant_edges[:1], quant_bins[:1] / total_count, 
       width=qwidth, fc=[0.3, 0.65, 0.3], alpha=0.5)
ax.bar(quant_edges[1:-1], np.cumsum(quant_bins[1:]) / total_count, 
       width=qwidth, fc=[0.3, 0.3, .75], alpha=0.5)
ax.plot(soft_edges[:-1], np.cumsum(soft_true_bins) / total_count, c=[0.2, 0.55, 0.2])
ax.plot(soft_edges[:-1], np.cumsum(soft_false_bins) / total_count, c=[0.2, 0.2, .65])
ax.set_xlabel("Distance")
ax.set_ylabel("Data Ratio")
ax.yaxis.labelpad = 15
fig.savefig("/Users/ejhumphrey/Dropbox/NYU/dl4mir/ejh-dissertation/src/6/figures/quant_error.pdf")
```

```python
post_fmt = "/Volumes/megatron/dl4mir/guitar/outputs/XL_weighted/0.125/{0}/posteriors/test.hdf5"
fret_fmt = "/Volumes/megatron/dl4mir/guitar/outputs/XL_weighted/0.125/{0}/fretboard/test.hdf5"
near_misses = []
for n in folds:
    post_stash = biggie.Stash(post_fmt.format(n))
    fret_stash = biggie.Stash(fret_fmt.format(n))
    for k in post_stash.keys():
        entity = post_stash.get(k)
        entity.fretboard = fret_stash.get(k).fretboard
        class_idx = vocab.label_to_index(entity.chord_labels)
        valid_idx = np.array([_ is not None for _ in class_idx])
        ref = templates[class_idx[valid_idx].astype(int), ...]
        est = templates[entity.posterior[valid_idx].argmax(axis=1)]
        sqerror = np.power(ref - est, 2.0).reshape(-1, 54).sum(axis=1)
        tf_idx = np.equal(sqerror, 0)
        
        quantized_dists.extend(qdist.tolist())
        sdist = np.sqrt(np.power(ref - entity.fretboard[valid_idx], 2.0).reshape(-1, 54).sum(axis=1))
        nm_idx = np.invert(tf_idx) * (sdist < 1.0)
        if nm_idx.sum():
            near_misses += [(entity.fretboard[nm_idx], ref[nm_idx], est[nm_idx], sdist[nm_idx])]
    if len(near_misses):
        break

    print "Finished {}".format(n)    
```
