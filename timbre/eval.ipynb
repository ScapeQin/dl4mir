```python
import glob
import json
import numpy as np
import tabulate
cfgs = 'c5', 'c8', 'c12', 'c24'
margins = '0.0', '0.25', '0.5'
splits = 'train', 'valid', 'test'
```

```python
# Combine results over folds

def merge_stats(stats):
    accum = dict()
    for res in stats:
        for k, v in res.iteritems():
            if k == 'class_labels':
                continue
            if not k in accum:
                accum[k] = []

            accum[k] += [v]
    for k in accum:
        accum[k] = np.asarray(accum[k])
        
    accum['class_labels'] = stats[0]['class_labels']
    return accum
```

```python
stats = dict()
for c in 'c5', 'c8', 'c12', 'c24':
    stats[c] = dict()
    for m in '0.0', '0.25', '0.5':
        stats[c][m] = merge_stats(
            [json.load(open(f)) 
             for f in glob.glob(fmt.format(config=c, margin=m))])
```

```python
fmt = "/Volumes/megatron/dl4mir/timbre_sim/results/pca-lda/256/{config}/*/stats.json"
lda_stats = dict()
for c in cfgs:
    lda_stats[c] = merge_stats([json.load(open(f)) 
                                for f in glob.glob(fmt.format(config=c))])
    
```

```python
# NLSE stats
tables = dict()
for split in splits:
    tables[split] = []
    for m in margins:
        tables[split].append([m])
        for c in cfgs:
            mu, sig = [fx(stats[c][m]['{0}_score'.format(split)]) 
                       for fx in np.mean, np.std]
            tables[split][-1].append("${0:.2f}\pm{1:.2f}$".format(mu*100, sig*100))

```

```python
for split in splits:
    print "%{0}\n{1}\n".format(split, 
                              tabulate.tabulate(tables[split], 
                                                headers=cfgs,
                                                tablefmt='plain'))
```
